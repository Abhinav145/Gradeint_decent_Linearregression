```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Making a linear dataset

```python
from sklearn.datasets import make_regression
X,y=make_regression( n_samples=4, n_features=1, n_informative=1,n_targets=1,noise=80,random_state=13)
```

 # E.D.A

```python
plt.scatter(X,y)
plt.show()
```

# Linear regression using O.L.S

```python
from sklearn.linear_model import LinearRegression
```

```python
clf=LinearRegression()
```

```python
clf.fit(X,y)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,clf.predict(X),color='red')
plt.show()
```

# lets us apply gradent decent assuming slope is constant m=78.35063668
# and lets us assume starting value for intercept is b=0


```python
y_predict=((78.35063668*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,clf.predict(X),color='red',label='OLS')
plt.plot(X,y_predict,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35063668
b=0
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

# lets take learning rate lr=0.1

```python
lr=0.1
step_size=loss_slope*lr
```

# calculating new intercept b

```python
b=b-step_size
```

```python
b
```

```python
y_predict1=((78.35063668*X)+b)
plt.scatter(X,y)
plt.plot(X,clf.predict(X),color='red')
plt.plot(X,y_predict,color='gray')
plt.plot(X,y_predict1,color='green')
```

# iteration 2

```python
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*lr
```

```python
# calculating new b
b=b-step_size
b
```

```python
y_predict2=((78.35063668*X)+b)
plt.scatter(X,y)
plt.plot(X,clf.predict(X),color='red')
plt.plot(X,y_predict,color='gray')
plt.plot(X,y_predict1,color='green')
plt.plot(X,y_predict2,color='green')
```
